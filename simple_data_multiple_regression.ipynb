# 2.Analysis Execution

```python
# purpose variable
y = df['魅力度']

#drop explanatory variable
X = df.drop(columns=['魅力度'])


## Adding constant (intercept) term to the data
import statsmodels.api as sm
X_ = sm.add_constant(X)

# Create an Ordinary Least Squares (OLS) regression model
model = sm.OLS(y, X_)

#execution analysis
result = model.fit()

#summary table
summary = result.summary()
summary
```

```python
#save as html
html = "<div>" + summary.as_html() + "</div>"

with open("multiple_regression.html","w",encoding="utf-8") as f:
    f.write(html)
```

```python
#save in pickle
import pickle
with open("pickle_regression.pkl", "wb") as f:
    pickle.dump(result,f)
```
